```python
import numpy as np
import networkx as nx

l = [(5, 7), (7, 1), (7, 2), (1, 4), (1, 6)]
G = nx.Graph()
G.add_edges_from(l)
nx.draw(G, with_labels=True)
```

```python
l = [("1", "2", 3.), 
     ("2", "3", 2.5), 
     ("3", "4", 1.5), 
     ("3", "5", 4.), 
     ("1", "6", 5.), 
     ("6", "7", 2.), 
     ("6", "8", 3.), 
     ("9", "8", 5.)]

G = nx.Graph()
G.add_weighted_edges_from(l)
labels = nx.get_edge_attributes(G, "weight")
nx.draw(G, with_labels=True)
```

```python
feature_vector = np.array([1, 7, 22])
G.add_node(100, v=feature_vector, label="n1") # New node
G.add_edge("1", 100, weight=8) # New edge
pos = nx.spring_layout(G)
nx.draw(G, pos=pos, with_labels=True)
```

```python
nx.draw(G, pos=pos, with_labels=True)

# Create edge labels
labels = {e: str(G.get_edge_data(e[0], e[1])["weight"]) for e in G.edges}
labels = {e: str(G.get_edge_data(*e)["weight"]) for e in G.edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=8);
```
