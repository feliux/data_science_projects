```python
from pyspark import SparkContext
from pyspark.sql import SQLContext, SparkSession
from pyspark.sql.types import *

from scipy.io import arff
import pandas as pd

#import numpy as np
#import seaborn as sb
#import matplotlib.pyplot as plt
#%matplotlib inline

data = arff.loadarff("../data/Phishing_Legitimate_full.arff")
df = pd.DataFrame(data[0])

#df.head()
#df.tail()

sc = SparkContext()
spark = SparkSession(sc)
sqlc = SQLContext(sc)

sc.version
```

```python
df.loc[(df["CLASS_LABEL"] == b"1"), "CLASS_LABEL"] = 1
df.loc[(df["CLASS_LABEL"] == b"0"), "CLASS_LABEL"] = 0
df.head()
```

```python
df.tail()
```

```python
d_schema = StructType([StructField("NumDots",DoubleType(),True),
                       StructField("SubdomainLevel",DoubleType(),True),
                       StructField("PathLevel",DoubleType(),True),
                       StructField("UrlLength",DoubleType(),True),
                       StructField("NumDash",DoubleType(),True),
                       StructField("NumDashInHostname",DoubleType(),True),
                       StructField("AtSymbol",DoubleType(),True),
                       StructField("TildeSymbol",DoubleType(),True),
                       StructField("NumUnderscore",DoubleType(),True), 
                       StructField("NumPercent",DoubleType(),True), 
                       StructField("NumQueryComponents",DoubleType(),True),
                       StructField("NumAmpersand",DoubleType(),True),
                       StructField("NumHash",DoubleType(),True),
                       StructField("NumNumericChars",DoubleType(),True),
                       StructField("NoHttps",DoubleType(),True),
                       StructField("RandomString",DoubleType(),True),
                       StructField("IpAddress",DoubleType(),True),
                       StructField("DomainInSubdomains",DoubleType(),True),
                       StructField("DomainInPaths",DoubleType(),True),
                       StructField("HttpsInHostname",DoubleType(),True),
                       StructField("HostnameLength",DoubleType(),True),
                       StructField("PathLength",DoubleType(),True),
                       StructField("QueryLength",DoubleType(),True),
                       StructField("DoubleSlashInPath",DoubleType(),True),
                       StructField("NumSensitiveWords",DoubleType(),True),
                       StructField("EmbeddedBrandName",DoubleType(),True),
                       StructField("PctExtHyperlinks",DoubleType(),True),
                       StructField("PctExtResourceUrls",DoubleType(),True),
                       StructField("ExtFavicon",DoubleType(),True),
                       StructField("InsecureForms",DoubleType(),True),
                       StructField("RelativeFormAction",DoubleType(),True),
                       StructField("ExtFormAction",DoubleType(),True),
                       StructField("AbnormalFormAction",DoubleType(),True),
                       StructField("PctNullSelfRedirectHyperlinks",DoubleType(),True),
                       StructField("FrequentDomainNameMismatch",DoubleType(),True),
                       StructField("FakeLinkInStatusBar",DoubleType(),True),
                       StructField("RightClickDisabled",DoubleType(),True),
                       StructField("PopUpWindow",DoubleType(),True),
                       StructField("SubmitInfoToEmail",DoubleType(),True),
                       StructField("IframeOrFrame",DoubleType(),True),
                       StructField("MissingTitle",DoubleType(),True),
                       StructField("ImagesOnlyInForm",DoubleType(),True),
                       StructField("SubdomainLevelRT",DoubleType(),True),
                       StructField("UrlLengthRT",DoubleType(),True),
                       StructField("PctExtResourceUrlsRT",DoubleType(),True),
                       StructField("AbnormalExtFormActionR",DoubleType(),True),
                       StructField("ExtMetaScriptLinkRT",DoubleType(),True),
                       StructField("PctExtNullSelfRedirectHyperlinksRT",DoubleType(),True),
                       StructField("CLASS_LABEL",StringType(),True)])

data = sqlc.createDataFrame(df, d_schema)
#data.dtypes
```

```python
sqlc.registerDataFrameAsTable(data, "bd")

bd2 = sqlc.sql("select NumDots, SubdomainLevel, PathLevel, InsecureForms, HttpsInHostname, UrlLength, CLASS_LABEL \
            from bd ")

sqlc.registerDataFrameAsTable(bd2, "bd2")
bd2.show(3)
```

```python
dummy = bd2.withColumn("YES",(bd2.CLASS_LABEL == 1)).withColumn("NO",(bd2.CLASS_LABEL == 0))
dummy.select("YES", "NO").toPandas().head(3)
```

```python
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.feature import VectorAssembler as VA
from pyspark.ml.feature import StringIndexer
from pyspark.ml.evaluation import BinaryClassificationEvaluator as BCE
from pyspark.sql.functions import col
#from pyspark.sql.functions import log

from time import time

# En RandomForestClassifier debemos transformar la variable objetivo a tipo double

a1 = VA(
    inputCols=[
        "NumDots",
        "SubdomainLevel",
        "PathLevel",
        "InsecureForms",
        "HttpsInHostname",
        "UrlLength"
    ],
    outputCol="features"
)

bd6 = a1.transform(bd2).select(col("CLASS_LABEL").cast("double").alias("label"),"features")

stringIndexer = StringIndexer(inputCol="label", outputCol="indexed")
sI = stringIndexer.fit(bd6)
bd6 = sI.transform(bd6)

splits = bd6.randomSplit([0.7, 0.3], 1234)
train = splits[0]
test = splits[1]

start_time = time()

rfc = RandomForestClassifier(
    labelCol="indexed",
    featuresCol="features",
    numTrees=500,
    maxDepth=5,
    seed = 1,
    impurity="gini",
    featureSubsetStrategy="sqrt"
)
     
rfc_model = rfc.fit(train)

end_time = time()
elapsed_time = end_time - start_time
print("Tiempo de entrenamiento: %.3f seconds" % elapsed_time)

predict_train = rfc_model.transform(train)
predict_test = rfc_model.transform(test)
predict_test.show(5)

ev = BCE(rawPredictionCol="rawPrediction", labelCol="indexed")
print("areaUnderROC test: {0}".format(ev.evaluate(predict_test, {ev.metricName: "areaUnderROC"})))
print("areaUnderROC training: {0}".format(ev.evaluate(predict_train, {ev.metricName: "areaUnderROC"})))
```
